```python
import numpy as np
import torch
from torchvision import datasets, models, transforms

data = torch.randn((1,3,4,4))
mean_tensor = torch.from_numpy(np.array([0.485, 0.456, 0.406]))
std_tensor = torch.from_numpy(np.array([0.229, 0.224, 0.225]))


transforms_normalization = transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])

a = data.clone()
a[:, 0, :, :] -= mean_tensor[0]
a[:, 0, :, :] /= std_tensor[0]
a[:, 1, :, :] -= mean_tensor[1]
a[:, 1, :, :] /= std_tensor[1]
a[:, 2, :, :] -= mean_tensor[2]
a[:, 2, :, :] /= std_tensor[2]
print(a)

b = data.clone()
b=transforms_normalization(b)
print(b)
```
